```python
data = pd.read_csv('tweets.csv')
data.head()
```

```python
print('Number of rows:',data.shape[0])
print('Number of columns:',data.shape[1])
print('\n')
data.info()
```

```python
data = data[['tweet_id','airline_sentiment','airline_sentiment_confidence','text']]
data
```

```python
# Drop rows with missing values
data.dropna(axis=0)

# Drop columns with missing values 
data.dropna(axis=1)

data = data.reset_index(drop=True)
```

```python
print('Number of rows:',data.shape[0])
print('Number of columns:',data.shape[1])
print('\n')
data.info()
data.describe(include='all')
```

```python
data.sample(1)
```

```python
le = LabelEncoder() 
data['airline_sentiment']= le.fit_transform(data['airline_sentiment']) 
data
```

```python
plt.figure(figsize=(9, 8))
sns.countplot(x='airline_sentiment',data=data)
```

```python
data['text'] = data.text.apply(lambda x: x.replace('@VirginAmerica ',''))
data.head()
```

```python
plt.figure(figsize=(9, 8))
sns.scatterplot(y='airline_sentiment_confidence',x='airline_sentiment', data=data)
```

```python
unlabeled = data[data.airline_sentiment_confidence < 0.6]
# unlabeled['airline_sentiment'] = unlabeled['airline_sentiment'].apply(lambda x: None)
unlabeled
```

```python
plt.figure(figsize=(9, 8))
sns.countplot(x='airline_sentiment',data=unlabeled)
```

```python
labeled = data[data.airline_sentiment_confidence >= 0.6]
labeled
```

```python
plt.figure(figsize=(9, 8))
sns.countplot(x='airline_sentiment',data=labeled)
```

```python
from sklearn.model_selection import train_test_split

df_trn, df_val = train_test_split(labeled, stratify = labeled['airline_sentiment'], test_size = 0.1, random_state = 5)
df_trn.shape, df_val.shape
```
